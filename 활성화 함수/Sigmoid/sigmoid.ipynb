```python
#   
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

#     ( 2 )
# n_informative=2, n_redundant=0, n_repeated=0   
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2,
                            n_informative=2, n_redundant=0, n_repeated=0,
                            random_state=42)

#     
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#  
plt.figure(figsize=(8, 6))
#  0  1     
sns.scatterplot(x=X_train[:, 0], y=X_train[:, 1], hue=y_train, palette='Set1', alpha=0.7)
#     
plt.title('   (2D )')
plt.xlabel(' 1')
plt.ylabel(' 2')
plt.legend(title='')
plt.show()

#  
model = Sequential()

#  (2 )
model.add(Dense(32, input_dim=2, activation='relu'))

# 
model.add(Dense(16, activation='relu'))

#  (    )
model.add(Dense(1, activation='sigmoid'))

#   (    , Adam  )
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#   (     )
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

#    
#   
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label=' ')
plt.plot(history.history['val_loss'], label=' ')
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.legend()

#  
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label=' ')
plt.plot(history.history['val_accuracy'], label=' ')
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.legend()

plt.tight_layout()
plt.show()

#  
loss, accuracy = model.evaluate(X_test, y_test)
print(f" : {loss:.4f},  : {accuracy:.4f}")
```
