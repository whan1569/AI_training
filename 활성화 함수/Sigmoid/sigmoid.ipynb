```python
# 필요한 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

# 이진 분류용 데이터셋 생성 (특성 2개로 제한)
# n_informative=2, n_redundant=0, n_repeated=0으로 설정하여 오류를 방지
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2,
                            n_informative=2, n_redundant=0, n_repeated=0,
                            random_state=42)

# 훈련 데이터와 테스트 데이터로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 데이터셋 시각화
plt.figure(figsize=(8, 6))
# 클래스 0과 클래스 1의 데이터 포인트를 다른 색으로 표시
sns.scatterplot(x=X_train[:, 0], y=X_train[:, 1], hue=y_train, palette='Set1', alpha=0.7)
# 그래프 제목 및 레이블 설정
plt.title('훈련 데이터셋 시각화 (2D 특성)')
plt.xlabel('특성 1')
plt.ylabel('특성 2')
plt.legend(title='클래스')
plt.show()

# 모델 생성
model = Sequential()

# 입력층 (2개의 특성)
model.add(Dense(32, input_dim=2, activation='relu'))

# 은닉층
model.add(Dense(16, activation='relu'))

# 출력층 (이진 분류이므로 시그모이드 함수 사용)
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일 (이진 교차 엔트로피 손실 함수, Adam 옵티마이저 사용)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 훈련 (훈련 진행 중 손실값과 정확도를 기록)
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# 훈련과 검증 성능 시각화
# 손실 값 시각화
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='훈련 손실')
plt.plot(history.history['val_loss'], label='검증 손실')
plt.title('훈련과 검증 손실')
plt.xlabel('에포크')
plt.ylabel('손실')
plt.legend()

# 정확도 시각화
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='훈련 정확도')
plt.plot(history.history['val_accuracy'], label='검증 정확도')
plt.title('훈련과 검증 정확도')
plt.xlabel('에포크')
plt.ylabel('정확도')
plt.legend()

plt.tight_layout()
plt.show()

# 모델 평가
loss, accuracy = model.evaluate(X_test, y_test)
print(f"테스트 손실: {loss:.4f}, 테스트 정확도: {accuracy:.4f}")
```
